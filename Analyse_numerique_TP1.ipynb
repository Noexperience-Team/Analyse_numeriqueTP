```python
import matplotlib.pyplot as plt
from numpy import linspace
import numpy as np
```

```python
def Horner(T,Z):
    n=len(T)
    result=T[0]
    for i in T[1:]:
        result=result*Z+i
        
    return result
```

```python
def p(x):
    return (x-1)**6
def F(x):
    return x**6-6*x**5+15*x**4-20*x**3+15*x**2-6*x+1
```

```python
T=[1,-6,15,-20,15,-6,1]
X=np.arange(0.998,1.002,0.0001)
ff=[]
pp=[]
for i in X:
    f=(Horner(T,i))
    pp.append(p(i))
    ff.append(F(i))
    print(f)
plt.plot(X,ff,'r',X,pp,'b*')
plt.show()
```

```python

def ploting(x=False):
    T=linspace(-2,2,41)
    z=np.zeros(41)
    plt.title("solution dans [-2:2] ");
    plt.xlabel("x")
    plt.ylabel ("f(x)")
    plt.grid(True)
    plt.plot(T,f(T))
    plt.plot(T,z,'y-')
    if x:
        plt.scatter(x,f(x),color='r')
    plt.show()
```

```python
def f(x):
    y=x**3+x**2-3*x-3
    
    return y
```

```python
ploting()
```

## Methode de dichotomie 
![alt text](https://upload.wikimedia.org/wikipedia/commons/8/8c/Bisection_method.svg)

```python
def dicho(f,a,b,tol):
    if (abs(a-b)<=tol) or f((a+b)/2)==0:
        m=(a+b)/2
    elif ((f((a+b)/2))*f(a))<0:
        m=dicho(f,a,(a+b)/2,tol)
    else:
        m=dicho(f,(a+b)/2,b,tol)
    return m

```

```python
point=dicho(f,-2,-1,0.00001)
ploting(point)
```
