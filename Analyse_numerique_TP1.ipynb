```python
import matplotlib.pyplot as plt
from numpy import linspace
```

```python

def ploting(x=False):
    zero=[0]*41
    T=linspace(-2,2,41)
    plt.title("solution dans [-2:2] ");
    plt.xlabel("x")
    plt.ylabel ("f(x)")
    plt.grid(True)
    plt.plot(T,f(T))
    plt.plot(T,zero,color='y')
    if x:
        plt.scatter(x,f(x),color='r')
    plt.show()
```

```python
def f(x):
    y=x**3+x**2-3*x-3
    
    return y
```

```python
ploting()
```

## Methode de dichotomie 
![alt text](https://upload.wikimedia.org/wikipedia/commons/8/8c/Bisection_method.svg)

```python
def dicho(f,a,b,tol):
    if (abs(a-b)<=tol) or f((a+b)/2)==0:
        m=(a+b)/2
    elif ((f((a+b)/2))*f(a))<0:
        m=dicho(f,a,(a+b)/2,tol)
    else:
        m=dicho(f,(a+b)/2,b,tol)
    return m

```

```python
point=dicho(f,-2,2,0.00001)
ploting(point)
print(point)
```

```python
def dichow(f,a,b,tol,count=0):
    if (abs(a-b)<=tol) or f((a+b)/2)==0:
        m=(a+b)/2
    elif ((f((a+b)/2))*f(a))<0:
        count+=1
        m,count=dichow(f,a,(a+b)/2,tol,count)
        
    else:
        count+=1
        m,count=dichow(f,(a+b)/2,b,tol,count)
    return (m,count)
```

```python
(point,count)=dichow(f,0,2,0.0001)
ploting(point)

print('le point est {},et elle est obtenu apres {} execution '.format(point,count))
```
