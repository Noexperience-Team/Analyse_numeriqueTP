```python
import matplotlib.pyplot as plt
from numpy import linspace
import numpy as np
```

```python
def Horner(T,Z):
    n=len(T)
    result=T[0]
    for i in T[1:]:
        result=result*Z+i
        
    return result
```

```python
def p(x):
    return (x-1)**6
def F(x):
    return x**6-6*x**5+15*x**4-20*x**3+15*x**2-6*x+1
```

```python
T=[1,-6,15,-20,15,-6,1]
X=np.arange(0.998,1.002,0.0001)
ff=[]
pp=[]
for i in X:
    f=(Horner(T,i))
    pp.append(p(i))
    ff.append(F(i))
    print(f)
plt.plot(X,ff,'r',X,pp,'b*')
plt.show()
```

```python

def ploting(f,x=False):
    T=linspace(-2,2,41)
    z=np.zeros(41)
    plt.title("solution dans [-2:2] ");
    plt.xlabel("x")
    plt.ylabel ("f(x)")
    plt.grid(True)
    plt.plot(T,f(T))
    plt.plot(T,z,'y-')
    if x:
        plt.scatter(x,f(x),color='r')
    
    plt.show()
```

# EXERCICE 2:
On considère l'équation :$$ f(x)=x^3+x^2-3x-3=0$$
1) Dessiner la courbe de $f$ dans l'intervalle $[-2;2]$, puis trouver des intervalles convenables pour appliquer la méthode de >bisection.

2) Utiliser un programme dicho.m qui permettra de trouver les solutions de l'équation $f(x)=0$ avec $eps=0.001$.

3) Modifier le programme dicho.m pour qu'il donne le nombre d'itération nécessaire pour avoir la solution à $eps$ près.



```python
def f(x):
    y=x**3+x**2-3*x-3
    
    return y
```

```python
ploting(f)
```

```python
def dicho(f,a,b,tol):
    if (abs(a-b)<=tol) or f((a+b)/2)==0:
        m=(a+b)/2
    elif ((f((a+b)/2))*f(a))<0:
        m=dicho(f,a,(a+b)/2,tol)
    else:
        m=dicho(f,(a+b)/2,b,tol)
    return m

```

```python
point=dicho(f,-2,-1,0.001)
ploting(f,point)
```

```python
def dichow(f,a,b,tol,x=0):
    if (abs(a-b)<=tol) or f((a+b)/2)==0:
        m=(a+b)/2
        n=x
    elif ((f((a+b)/2))*f(a))<0:
        m,n=dichow(f,a,(a+b)/2,tol,x+1)
        
    else:
        m,n=dichow(f,(a+b)/2,b,tol,x+1)
           
    return m,n
```

```python
print(dichow(f,-2,-1.5,0.001))
```

```python
interval=[[-2,-1.5],[-1.5,-0.5],[1.5,2]]
for u in interval:
    print(dichow(f,u[0],u[1],10**(-3)))
```

### methode de point fix :fonction

```python
def pointFix(g,x0,eps,nmax):
    zero=x0
    for niter in range(1,nmax+1):
        x=zero
        zero=g(x)
        erreur=abs(zero-x)
        if erreur<eps: 
            return zero,niter,erreur
    return print('pas de convergence avec le nombre d''iteration indique')
```

```python
print(pointFix(g1,1.5,0.001,50))
```

```python

print(pointFix(g2,1.5,0.001,50))
```

```python
print(pointFix(g3,1.5,0.001,7))
```

## EXERCICE 4:
1) On considère la fonction $f(x)=e^{-x}-x$. Ecrire puis appliquer une fonction Matlab que l'on appellera newton.m à la >fonction $f$, en prenant $x0=0$, $eps=0.001$ .

2) Montrer que l'équation $f(x)=0$ admet une solution unique que l'on peut obtenir en utilisant la méthode de Newton à partir >de $x0=1$.



```python
import math
def f4(x):
    y=(math.exp(-1*x))-x
    return y
def fprime(x):
    return -1*math.exp(-1*x)-1
```

```python
ff4=np.vectorize(f4)
z=np.zeros(100)
t=np.linspace(0,1,100)
plt.plot(t,ff4(t))
plt
plt.grid(True)
plt.show()
```

```python

def newtonn(a,eps,f,fprime):
    x1=a
    x2=x1-f(x1)/fprime(x1);

    while abs(x1-x2)>eps:
         x1=x2
         x2=x1-f(x1)/fprime(x1)

    return x2
```

```python
print(newtonn(1,0.001,f4,fprime))
```
